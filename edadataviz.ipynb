```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create catplot for FlightNumber vs LaunchSite
sns.catplot(
    x="FlightNumber",  # x-axis: Flight Number
    y="LaunchSite",    # y-axis: Launch Site
    hue="Class",       # Color by mission outcome
    data=df,
    kind="strip",      # Scatter plot style
    aspect=3,          # Width-to-height ratio
    height=6,          # Height in inches
    palette="Set2",    # Color palette
    s=50,              # Marker size
    edgecolor="black"  # Marker edge color
)

# Customize the plot
plt.title("Flight Number vs Launch Site by Mission Outcome", fontsize=16, pad=20)
plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Launch Site", fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Calculate success rate for each orbit type
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()
orbit_success.columns = ['Orbit', 'SuccessRate']

# Sort by success rate (descending)
orbit_success = orbit_success.sort_values('SuccessRate', ascending=False)

# Create bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(orbit_success['Orbit'], orbit_success['SuccessRate'], 
               color='skyblue', edgecolor='black')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.2%}', ha='center', va='bottom', fontsize=10)

# Customize the plot
plt.title('Success Rate by Orbit Type', fontsize=16, pad=20)
plt.xlabel('Orbit Type', fontsize=14)
plt.ylabel('Success Rate', fontsize=14)
plt.ylim(0, 1.1)  # Set y-axis limit to accommodate labels
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Function to extract years from the date
def Extract_year():
    year = []
    for i in df["Date"]:
        year.append(str(i).split("-")[0])
    return year

# Apply the function
df['Year'] = Extract_year()

# Calculate yearly success rate
yearly_success = df.groupby('Year')['Class'].mean().reset_index()
yearly_success.columns = ['Year', 'SuccessRate']

# Convert Year to numeric for proper sorting
yearly_success['Year'] = yearly_success['Year'].astype(int)
yearly_success = yearly_success.sort_values('Year')

# Create line chart
plt.figure(figsize=(12, 6))
plt.plot(yearly_success['Year'], yearly_success['SuccessRate'], 
         marker='o', markersize=8, linewidth=3, color='steelblue')

# Add data point labels
for x, y in zip(yearly_success['Year'], yearly_success['SuccessRate']):
    plt.text(x, y + 0.02, f'{y:.1%}', ha='center', va='bottom', fontsize=10)

# Customize the plot
plt.title('Yearly Launch Success Trend', fontsize=16, pad=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Success Rate', fontsize=14)
plt.ylim(0, 1.1)  # Set y-axis limits
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(yearly_success['Year'])
plt.tight_layout()
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
import pandas as pd

# First, let's check what categorical columns we have
print("Original DataFrame columns:")
print(df.columns.tolist())
print("\nDataFrame shape:", df.shape)

# Create the features dataframe
features = df.copy()

# Identify categorical columns to encode
categorical_columns = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# Check which of these columns actually exist in your dataframe
existing_categorical = [col for col in categorical_columns if col in features.columns]
print("\nCategorical columns to encode:", existing_categorical)

# Apply One-Hot Encoding using get_dummies
features_one_hot = pd.get_dummies(
    features, 
    columns=existing_categorical, 
    prefix=existing_categorical,
    dtype=int
)

print("\nEncoded DataFrame shape:", features_one_hot.shape)
print("\nFirst 5 rows of encoded features:")
print(features_one_hot.head())
```
